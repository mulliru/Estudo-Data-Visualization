# Aula 1
Fazendo as primeiras visualizações

```python
import pandas as pd
```

```python
df = pd.read_csv('.\curso-matplotlib-preparando_ambiente\curso-matplotlib-preparando_ambiente\monitoramento_tempo.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df["data"] = pd.to_datetime(df["data"]) #convertendo a coluna data de object (string) para "datetime" (data e hora)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df["data"], df["temperatura"])
```

```python
plt.figure(figsize=(15,8))
plt.plot(df["data"], df["temperatura"])
```

```python
plt.figure(figsize=(15,8))
plt.plot(df["data"], df["temperatura"])
plt.title("temperatura no tempo")
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0, 1,1])
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0, 1,1])
eixo.plot(df["data"], df["temperatura"])

eixo.set_title("temperatura no momento")
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0, 1,1])
eixo.plot(df["data"], df["temperatura"])

eixo.set_title("temperatura no momento") #adicionando título ao gráfico
eixo.set_ylabel("temperatura") #adicionando legenda ao eixo y
eixo.set_xlabel("data") #adicionando legenda ao eixo x
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0, 1,1])
eixo.plot(df["data"], df["temperatura"])

eixo.set_title("temperatura no momento", fontsize=25) #colocando o tamanho da fonte
eixo.set_ylabel("temperatura", fontsize=20) #colocando o tamanho da fonte
eixo.set_xlabel("data", fontsize=20) #colocando o tamanho da fonte
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0, 1,1])
eixo.plot(df["data"], df["temperatura"])

eixo.set_title("temperatura no momento", fontsize=25)
eixo.set_ylabel("temperatura", fontsize=20)
eixo.set_xlabel("data", fontsize=20)
eixo.legend(["temperatura"], loc="lower right", fontsize=15) #adicionando legenda no canto inferior direito
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0, 1,1])
eixo.plot(df["data"], df["temperatura"], color="green") #mudando a cor

eixo.set_title("temperatura no momento", fontsize=25)
eixo.set_ylabel("temperatura", fontsize=20)
eixo.set_xlabel("data", fontsize=20)
eixo.legend(["temperatura"], loc="lower right", fontsize=15) 
```

# Aula 2

Customizando nossas visualizações

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0, 1,1])
eixo.plot(df["data"], df["temperatura"], color="green", lw = 4) #mudando a espessura da linha

eixo.set_title("temperatura no momento", fontsize=25)
eixo.set_ylabel("temperatura", fontsize=20)
eixo.set_xlabel("data", fontsize=20)
eixo.legend(["temperatura"], loc="lower right", fontsize=15) 
```
